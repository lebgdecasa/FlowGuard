```python
# Create binary indicators for zero values
df['has_response'] = (df['resp_bytes'] > 0).astype(int)
df['sent_data'] = (df['orig_bytes'] > 0).astype(int)
df['has_resp_pkts'] = (df['resp_pkts'] > 0).astype(int)

# Now you can drop the original highly-zero columns
df = df.drop(['resp_bytes', 'resp_pkts', 'resp_ip_bytes'], axis=1)
```

```python
# Drop the redundant features (keep the simpler ones)
df = df.drop(['orig_ip_bytes'], axis=1)  # Keep orig_pkts instead
# We already dropped resp_ip_bytes above
```

```python
# First, frequency encode the history column
history_freq = df['history'].value_counts().to_dict()
df['history_freq'] = df['history'].map(history_freq)

# Then extract key patterns
df['history_length'] = df['history'].str.len()
df['history_has_S'] = df['history'].str.contains('S').astype(int)
df['history_has_D'] = df['history'].str.contains('D').astype(int)
df['history_has_F'] = df['history'].str.contains('F').astype(int)
df['history_has_a'] = df['history'].str.contains('a').astype(int)
df['history_has_r'] = df['history'].str.contains('r').astype(int)

# Count unique characters in history
df['history_unique_chars'] = df['history'].apply(lambda x: len(set(x)))

# Now drop the original history column
df = df.drop(['history'], axis=1)
```

```python
# Step 4: Create New Features (Adjusted)

# 4.1 Create ratio features
print("Creating ratio features...")

# Since resp_bytes is dropped, we'll use the binary indicator logic
# If has_response is 0, then resp_bytes was 0, so bytes_ratio = 1
# If has_response is 1, we can't calculate exact ratio, so we'll skip this feature
# OR we can create a simpler version:

# Alternative approach - just use what we have
df['sent_data_ratio'] = df['sent_data']  # Already binary, represents if data was sent

# For packets, check if we still have resp_pkts
if 'resp_pkts' in df.columns:
    df['pkts_ratio'] = df['orig_pkts'] / (df['orig_pkts'] + df['resp_pkts'] + 1)
else:
    # Use binary indicator instead
    df['pkts_all_sent'] = ((df['orig_pkts'] > 0) & (df['has_resp_pkts'] == 0)).astype(int)

# 4.2 Log transform skewed features
print("Applying log transforms...")
df['duration_log'] = np.log1p(df['duration'])

# Check if orig_bytes still exists
if 'orig_bytes' in df.columns:
    df['orig_bytes_log'] = np.log1p(df['orig_bytes'])
else:
    print("orig_bytes already dropped")

# Log transform orig_pkts too
df['orig_pkts_log'] = np.log1p(df['orig_pkts'])

# Check the new features
print("\nNew features created successfully!")
```

```python
# Step 5: One-hot encode protocol

# First, let's check what protocols we have
print("Unique protocols in dataset:")
print(df['proto'].value_counts())

# One-hot encode protocol
print("\nOne-hot encoding protocol...")
df = pd.get_dummies(df, columns=['proto'], prefix='proto')

# Verify the encoding worked
print("\nNew protocol columns created:")
proto_cols = [col for col in df.columns if col.startswith('proto_')]
print(proto_cols)

# Check shape
print(f"\nDataset shape after protocol encoding: {df.shape}")
```

```python
# Step 6: Encode connection state

# First, let's examine connection states
print("Connection state distribution:")
conn_state_counts = df['conn_state'].value_counts()
print(conn_state_counts)
print(f"\nTotal unique connection states: {len(conn_state_counts)}")

# Show top 10 states with percentages
print("\nTop 10 connection states:")
for state, count in conn_state_counts.head(10).items():
    percentage = (count / len(df)) * 100
    print(f"{state}: {count:,} ({percentage:.2f}%)")

# For encoding, we'll keep top states and group rare ones
top_n = 10  # Adjust if needed
top_states = conn_state_counts.head(top_n).index.tolist()

# Create grouped column
df['conn_state_grouped'] = df['conn_state'].apply(
    lambda x: x if x in top_states else 'Other'
)

# One-hot encode
print("\nOne-hot encoding connection states...")
df = pd.get_dummies(df, columns=['conn_state_grouped'], prefix='conn_state')

# Drop original conn_state column
df = df.drop(['conn_state'], axis=1)

# Check results
print(f"\nDataset shape after connection state encoding: {df.shape}")
conn_state_cols = [col for col in df.columns if col.startswith('conn_state_')]
print(f"Connection state columns created: {len(conn_state_cols)}")
```

```python
# Step 7: Optional - Port number features (can be useful for Random Forest)

# Check port distributions first
print("Port number statistics:")
print("Origin ports:", df['id.orig_p'].describe())
print("\nResponse ports:", df['id.resp_p'].describe())

# Create useful port features
print("\nCreating port features...")

# Common service ports
common_ports = [80, 443, 22, 21, 25, 53, 123, 445, 3389, 3306, 8080, 8443]
df['orig_port_common'] = df['id.orig_p'].isin(common_ports).astype(int)
df['resp_port_common'] = df['id.resp_p'].isin(common_ports).astype(int)

# High vs low ports (>1024 are typically client ports)
df['orig_port_high'] = (df['id.orig_p'] > 1024).astype(int)
df['resp_port_high'] = (df['id.resp_p'] > 1024).astype(int)

# Step 8: Final cleanup - drop original columns we've transformed
columns_to_drop = ['duration', 'orig_bytes']  # We have log versions
if all(col in df.columns for col in columns_to_drop):
    df = df.drop(columns=columns_to_drop, axis=1)
    print(f"\nDropped original columns: {columns_to_drop}")

# Final check
print(f"\n=== FINAL DATASET SUMMARY ===")
print(f"Shape: {df.shape}")
print(f"Columns ({len(df.columns)}): {sorted(df.columns.tolist())}")
print(f"\nNull values: {df.isnull().sum().sum()}")
print(f"\nLabel distribution:")
print(df['label'].value_counts())
print(f"Malicious ratio: {(df['label'].sum() / len(df) * 100):.2f}%")
```

```python
# Split the data (70-30 split, stratified to maintain class distribution)
print("Splitting data into train and test sets...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y  # Important: maintains the 57/43 ratio in both sets
)

print(f"\nTrain set: {X_train.shape}")
print(f"Test set: {X_test.shape}")

# Verify stratification worked
print("\nTrain set class distribution:")
print(y_train.value_counts())
print(f"Train malicious ratio: {(y_train.sum() / len(y_train) * 100):.2f}%")

print("\nTest set class distribution:")
print(y_test.value_counts())
print(f"Test malicious ratio: {(y_test.sum() / len(y_test) * 100):.2f}%")
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select categorical columns to encode
categorical_cols = ['proto', 'conn_state', 'history']

# Fit OneHotEncoder on the categorical columns using X_train
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Fit and transform on X_train
encoded_train = encoder.fit_transform(X_train[categorical_cols])

# Only transform X_test
encoded_test = encoder.transform(X_test[categorical_cols])

# Get new column names
encoded_feature_names = encoder.get_feature_names_out(categorical_cols)

# Create DataFrames for encoded features
encoded_train_df = pd.DataFrame(encoded_train, columns=encoded_feature_names, index=X_train.index)
encoded_test_df = pd.DataFrame(encoded_test, columns=encoded_feature_names, index=X_test.index)
```

```python
from sklearn.preprocessing import RobustScaler

# Let's first check for outliers in a few features
print("Checking for outliers in key features:")
features_to_check = ['orig_pkts', 'duration', 'orig_bytes', 'resp_bytes', 'orig_ip_bytes',  'resp_pkts', 'resp_ip_bytes']
for feat in features_to_check:
    if feat in X_train.columns:
        col_idx = X_train.columns.get_loc(feat)
        q1 = np.percentile(X_train.iloc[:, col_idx], 25)
        q3 = np.percentile(X_train.iloc[:, col_idx], 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = ((X_train.iloc[:, col_idx] < lower_bound) | (X_train.iloc[:, col_idx] > upper_bound)).sum()
        print(f"{feat}: {outliers:,} outliers ({outliers/len(X_train)*100:.2f}%)")

# Use RobustScaler instead
print("\nScaling features with RobustScaler...")
scaler = RobustScaler()

# Fit on training data and transform
X_train_scaled_num = scaler.fit_transform(X_train[features_to_check])

# Transform test data
X_test_scaled_num = scaler.transform(X_test[features_to_check])

print("\nScaling complete!")
print(f"Scaled train shape: {X_train_scaled_num.shape}")
print(f"Scaled test shape: {X_test_scaled_num.shape}")
```

```python
df_X_train_scaled_num = pd.DataFrame(X_train_scaled_num, columns=features_to_check, index=X_train.index)
df_X_test_scaled_num = pd.DataFrame(X_test_scaled_num, columns=features_to_check, index=X_test.index)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
import time

print("=== RANDOM FOREST MODEL ===")

# Note: We'll use the ORIGINAL unscaled data for Random Forest
# RF doesn't need scaling and actually performs better without it

# Start with a reasonable RF configuration
print("\nTraining Random Forest with optimized parameters...")
rf_model = RandomForestClassifier(
    n_estimators=100,          # Start with 100 trees
    max_depth=20,              # Limit depth to prevent overfitting
    min_samples_split=50,      # Require 50 samples to split
    min_samples_leaf=20,       # At least 20 samples in leaf
    max_features='sqrt',       # sqrt(36) ≈ 6 features per split
    class_weight='balanced',   # Handle class imbalance
    random_state=42,
    n_jobs=-1,                 # Use all cores
    verbose=1                  # Show progress
)

# Train on original (unscaled) data
start_time = time.time()
rf_model.fit(X_train, y_train)
training_time = time.time() - start_time

print(f"\nTraining completed in {training_time:.2f} seconds")

# Make predictions
print("\nEvaluating on test set...")
y_pred_rf = rf_model.predict(X_test)
y_pred_proba_rf = rf_model.predict_proba(X_test)[:, 1]

# Calculate metrics
roc_auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
print(f"\nRandom Forest ROC-AUC: {roc_auc_rf:.4f}")
print(f"Baseline ROC-AUC: {roc_auc:.4f}")
print(f"Improvement: {(roc_auc_rf - roc_auc):.4f} ({((roc_auc_rf - roc_auc)/roc_auc)*100:.1f}%)")

# Detailed evaluation
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['Benign', 'Malicious']))

# Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(f"\nConfusion Matrix:")
print(f"              Predicted")
print(f"              Benign    Malicious")
print(f"Actual Benign    {cm_rf[0,0]:,}     {cm_rf[0,1]:,}")
print(f"      Malicious  {cm_rf[1,0]:,}     {cm_rf[1,1]:,}")

# Feature importance from Random Forest
feature_importance_rf = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\n=== TOP 15 FEATURES BY RANDOM FOREST ===")
print(feature_importance_rf.head(15).to_string(index=False))
```
